```python
import gensim.downloader as api
import numpy as np
from numpy.linalg import norm



glove = api.load("glove-wiki-gigaword-50")
```

```python
similarity1 = glove.similarity('king', 'queen')
similarity2 = glove.similarity('kiwi', 'apple')
similarity3 = glove.similarity('phone', 'apple')


print("similarty 1 " , similarity1)
print("similarty 2 " , similarity2)
print("similarty 3 " , similarity3)
```

```python
## sentence semanti similarity

def sentence_vector(sentence):
  words=sentence.lower().split()
  vectors = []

  for word in words:
    if word in glove:
      vectors.append(glove[word])

  return np.mean(vectors , axis=0)
```

```python
sentence1 = "I love machine learning"
sentence2 = "I love deep learning"
sentence3 = "The cat is sleeping"


vector1 = sentence_vector(sentence1)
vector2 = sentence_vector(sentence2)
vector3 = sentence_vector(sentence3)


def cosine_similarity(a, b):
    return np.dot(a , b)/(norm(a)* norm(b))

print("similarity(s1, s2)", cosine_similarity(vector1, vector2))
print("similarity(s2, s3)", cosine_similarity(vector2, vector3))
```
