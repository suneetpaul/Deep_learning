```python
import numpy as np
```

```python
# Sample Sentences
# sentence = 'I love my state Uttarakhand also known as Devbhoomi'
sentence = 'i come from a state called uttarakhand which is also known as devbhoomi'
sentence = "l love machine learning"
```

```python
#tokenization
words = sentence.split()
```

```python
#vocabulary
vocab = list(words)
vocab_size = len(vocab)
```

```python
#word to index mapping
word_to_index = {word: index for index, word in enumerate(vocab)}
#index to word mapping
index_to_word = {index: word for index, word in enumerate(vocab)}
```

```python
# generate trannig data (window size = 1)
window_size = 1
training_data = []
for i in range(window_size , len(words)- window_size):
    context = []
    for j in range(-window_size , window_size + 1):
        if j != 0:
            context.append(word_to_index[words[i +j]])

    target = word_to_index[words[i]]
    training_data.append((context , target))

# training_data = np.array(training_data)
print(training_data)
```

```python
## initialize weights
#size f word vector
embeddig_dim = 5

#input weight matrix

W1 = np.random.randn(vocab_size , embeddig_dim)

W2 = np.random.randn(embeddig_dim , vocab_size)

learning_rate = 0.01
```

```python
# softmax funtion

def softmax(x):
  exp_x = np.exp(x-np.max(x))
  return exp_x / exp_x.sum()
```

```python
# traning (basic grandient descent)

epochs = 1000

for epoch in range(epochs):
  loss =0

  for context , target in training_data:
    #forwad pss
    h = np.mean(W1[context], axis =0)
    u = np.dot(h,W2)
    y_pred = softmax(u)


    #loss( cross- entropy)

    loss -= np.log(y_pred[target])


    #backpropagation

    error = y_pred
    error[target] -=1

    #gradient for w2
    dw2 = np.outer(h,error)

    #gradient for w1
    dw1 = np.dot(W2,error) / len( context)


    #update w2
    W2 -= learning_rate *dw2

    #update w1 (only context words)
    for word in context:
      W1[word] -= learning_rate*dw1

  if epoch % 200 == 0:
    print(f"epoch{epoch} , loss:{loss:.4f}")
```

```python
print("\n word embedding:")

for word in vocab:
  print(word , '->' , W1[word_to_index[word]])
```
