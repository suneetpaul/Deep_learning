```python
import numpy as np
```

```python
sentence = "I machine learning"

word = sentence.split()

#vocabulary

vocab = list(set(word))
vocab_size = len(vocab)

word_to_index = {word:i for i, word in enumerate(vocab)}
index_to_word = {i:word for i, word in word_to_index.items()}
```

```python
##generate training data window size= 1


window_size = 1

traing_data = []

for i in range(window_size , len(word) - window_size):
  target = word_to_index[word[i]]

  for j in range(-window_size, window_size+1):
    if j != 0:
      context = word_to_index[word[i+j]]

      traing_data.append((target, context))


print(traing_data)
```

```python
##initalize weight

embedding_dim = 5
learning_rate = 0.01


#input weight (target word vector)

W1 = np.random.randn(vocab_size, embedding_dim)
#output weight

W2 = np.random.randn(embedding_dim , vocab_size)
```

```python
# softmax funtion

def softmax(x):
  exp_x = np.exp(x-np.max(x))
  return exp_x / exp_x.sum()
```

```python
# traning (basic grandient descent)

epochs = 1000

for epoch in range(epochs):
  loss =0

  for target, context in traing_data:
    #forwad pss
    h = W1[target]
    u = np.dot(h,W2)
    y_pred = softmax(u)


    #loss( cross- entropy)

    loss -= np.log(y_pred[context])


    #backpropagation

    error = y_pred
    error[target] -=1

    #gradient for w2
    dw2 = np.outer(h,error)

    #gradient for w1
    dw1 = np.dot(W2,error)


    #update w2
    W2 -= learning_rate *dw2

    #update w1 (only context words)

    W1[target] -= learning_rate*dw1

  if epoch % 200 == 0:
    print(f"epoch{epoch} , loss:{loss:.4f}")
```

```python
print("\n word embedding:")

for word in vocab:
  print(word , '->' , W1[word_to_index[word]])
```
