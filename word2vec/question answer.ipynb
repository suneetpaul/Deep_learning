```python
import re
import emoji
import demoji
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
```

```python
def clean_text(article):

  text = article.lower()

  text = re.sub(r'\d+', '', text)
  text = re.sub(r'\s+', ' ', text)
  text = emoji.replace_emoji(text, replace="")
  text = demoji.replace(text, "")

  words = nltk.word_tokenize(text)

  cleaned_words_list = []
  for word in words:
       if word not in stop_words:
           lemma = lemmatizer.lemmatize(word)
           cleaned_words_list.append(lemma)

  return " ".join(cleaned_words_list), cleaned_words_list
```

```python


clean_text(article)
```

```python
text_cleaned, cleaned_words_list = clean_text(article)
```

```python
for i in cleaned_words_list:
  if len(i) <= 1:
    print(i)
```

```python


from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
vectorizer = TfidfVectorizer()
```

```python
sentences = sent_tokenize(text_cleaned)
article_vectors = vectorizer.fit_transform(sentences)
```

```python

question = "unsupervised learning"

ques_cleaned_string, _ = clean_text(question)
ques_vector = vectorizer.transform([ques_cleaned_string])

similarity = cosine_similarity(ques_vector, article_vectors)
index = np.argsort(similarity[0])[::-1]

score = similarity[0][index[-1]]


#if score < 0.2:
#    return ( "Sorry, I couldn't find an answer.", score)

top_answers = []

for i in index:
    if similarity[0][i] > 0.2:
        top_answers.append((sentences[i], similarity[0][i]))

top_answers
```

```python

def answer(question):

    ques_cleaned_string, _ = clean_text(question)
    ques_vector = vectorizer.transform([ques_cleaned_string])

    similarity = cosine_similarity(ques_vector, article_vectors)
    index = np.argsort(similarity[0])[::-1]

    score = similarity[0][index[0]]


    if score < 0.2:
        return ( "Sorry, I couldn't find an answer.", score)

    top_answers = []

    for i in index:
        if similarity[0][i] > 0.2:
            top_answers.append((sentences[i], similarity[0][i]))

    return top_answers
```

```python
answer("unsupervised learning")
```

```python
while True:
    question = input("Enter your question (type 'exit' to stop): ")

    if question.lower() == "exit":
        print("Program ended.")
        break

    responses = answer(question)


    if responses and responses[0][1] == 0.0:
        print("Answer:", responses[0][0])
    else:

        for sentence in responses:

            print(f"Sentence: {sentence[0]}")
```
