```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
```

```python
document = [
    "Data science includes machine learning",
    "Machine learning algorithms build models",
    "Deep learning uses neural network",
    "Neural network are used in deep learning",
    "Football match was exciting",
    "The team won the football game"
]
```

```python
vectorize = TfidfVectorizer(stop_words='english')
X = vectorize.fit_transform(document)
```

convertr text to numerical form

```python
print( X)
```

```python
nmf = NMF(n_components=2 , random_state=42)
W = nmf.fit_transform(X)
H = nmf.components_
```

n_components = 2 ---> number of topic

W --> documten - topic matrix

H --> topic - word matrix

```python
## disply topic

feature_names = vectorize.get_feature_names_out()

for topic_idx , topic in enumerate(H):
    print("Topic %d:" % (topic_idx + 1))
    top_words = [ feature_names[i] for i in topic.argsort()[-6:]]

    print(top_words)
```

**NMF performs topic modling by factorizing a TF-IDF matrix into non - negative document - topic and topic- term matrices , producing interpretable topixs.**
