```python
import gensim
from gensim import corpora
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
document =[

    "Data science uses machine learning algorithms",
    "Machine learning is a subset of artificial intelligence",
    "Deep learning uses neural networks",
    "Neural networks are part of deep learning",
    "Python is popular for data science",
    "Statistics is important for data analysis",
    "Football match was exciting",
    "The team won the football game",
    "Players scored goals in the match",
    "The football tournament was successful"
]
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt_tab')

stop_word = set(stopwords.words('english'))

text = []

for doc in document:
    tokens = word_tokenize(doc.lower())
    tokens = [word for word in tokens if word.isalpha() and word not in stop_word]
    text.append(tokens)
```

```python
disctionary = corpora.Dictionary(text)
corpus = [disctionary.doc2bow(doc) for doc in text]
```

```python
lda_model = gensim.models.LdaModel(
    corpus = corpus,
    id2word = disctionary,
    num_topics = 3,
    passes = 10
)
```

```python
for topic in lda_model.print_topics():
    print(f"Topic {topic}")

```

```python
lda_model = gensim.models.LdaModel(
    corpus = corpus,
    id2word = disctionary,
    num_topics = 3,
    passes = 1
)
```

```python
for topic in lda_model.print_topics():
    print(f"Topic {topic}")
```
